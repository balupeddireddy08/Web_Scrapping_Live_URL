```python
# Retrieve and print the number of available CPU cores using the multiprocessing module and os.
import os
import multiprocessing

num_cores = multiprocessing.cpu_count()
print(f"Number of CPU cores available: {num_cores}")
```

```python
import requests
# Code to store the retrievable links from a given set of URLs.
# Define a function to fetch a URL
def fetch_url(url):
    try:
        page = requests.get(url, timeout=10)  # Timeout of 10 seconds
        page.raise_for_status()  # Check for HTTP errors
        print(url)
        return url, page.text
    except requests.exceptions.RequestException as e:
        print(e) # Print the exception if an error occurs
        return url, f"Error fetching {url}: {str(e)}"

if __name__ == "__main__":
    # Creating a multiprocessing Pool with the desired number of processes
    num_processes = multiprocessing.cpu_count()
    pool = multiprocessing.Pool(processes=num_processes)
    # Use the pool to concurrently fetch the URLs
    results = pool.map(fetch_url, URLS[:])
    # Close the pool to free up resources
    pool.close()
    pool.join()
    Updated_URLS = []
    for url, result in results:
        if not result.startswith("Error"):
            Updated_URLS.append(url)
        print(result)
```

```python
dict_url_category = {}
import re

# Regular expression pattern to extract domain names from the URLS
pattern = r"https://(www\.)?(\w+)\."

# Extract domain names from URLs
domain_names = [re.search(pattern, url).group(2) for url in Updated_URLS]
domain_names_urls = [(re.search(pattern, url).group(2), url) for url in Updated_URLS]

# Print the extracted domain names and URLS
print(domain_names)
print(domain_names_urls)
# This code uses the regular expression r"https://(www\.)?(\w+)\." to match the domain part of the URL. It captures both with and without "www" subdomains and extracts the primary domain name from each URL.
# The re.search() function is used to find the pattern in each URL, and group(2) is used to extract the second captured group, which contains the domain name.

print(set(list(domain_names)))
for i in set(list(domain_names)):
    dict_url_category[i] = []
```

```python
# Create a Dictionary that contains similar URLS
for website_name, url in domain_names_urls:
    dict_url_category[website_name].append(url)
# Create a Dictionary that URL as key and domain_name as value
dict_url = {}
for website_name, url in domain_names_urls:
    dict_url[url] = website_name
print(dict_url)
print(dict_url_category)
```

```python
# Code to retrieve the names of products from a website page.

from bs4 import BeautifulSoup
import requests
dict_heading = {}
for url in Updated_URLS:
    try:
        names = []
        page = requests.get(url, timeout=10)  #timeout of 10 seconds
        page.raise_for_status()  # Checking for HTTP errors
        soup = BeautifulSoup(page.content, "html.parser")
        # Perform website-specific HTML parsing based on the 'dict_url' mapping. (URL_domain_specific)
        if dict_url[url] in ['thestreet', 'twentytwowords', 'mightymoms', 'sailmagazine']:
            tags = soup.find_all('div', class_='item-recommend')
            for tag in tags:
                names.append(tag.h3.text)
            print(url)
            print(names)
            dict_heading[url] = names
        elif dict_url[url] in ['hollywoodlife']:
            tags = soup.find_all('article', class_='review-product')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))
            print(url)
            print(names)
            dict_heading[url] = names
        elif dict_url[url] in ['oldhouseonline']:
            tags = soup.find_all('div', class_='product-title-wrap')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))
            print(url)
            print(names)
            dict_heading[url] = names
        elif dict_url[url] in ['techjunkie']:
            tags = soup.find_all('div', class_='best-product-box')
            for tag in tags:
                if tag.h2.text.replace('\n', '').replace('\t', '').replace('\xa0', '').startswith("0"):
                    names.append(tag.h2.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))
            print(url)
            print(names)
            dict_heading[url] = names
        elif dict_url[url] in ['usmagazine']:
            tags = soup.select('div.article-content h3')
            product_names = [h3.text.strip() for h3 in tags]
            for product_name in product_names:
                if '  ' in product_name:
                    names.append(product_name)
            print(url)
            print(names)
            dict_heading[url] = names
        elif dict_url[url] in ['cuisineathome', 'woodsmith', 'gardengatemagazine']:
            tags = soup.find_all('div', class_='product-title-wrap')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))
            print(url)
            print(names)
            dict_heading[url] = names
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {str(e)}")
```

```python
# Code to retrieve the amazon links of products from a website page.

from bs4 import BeautifulSoup
import requests
dict_links = {}
for url in Updated_URLS:
    try:
        links = []
        page = requests.get(url, timeout=10)  #timeout of 10 seconds
        page.raise_for_status()  # Checking for HTTP errors
        soup = BeautifulSoup(page.content, "html.parser")
        if dict_url[url] in ['thestreet', 'twentytwowords', 'mightymoms', 'sailmagazine']:
            tags = soup.find_all('div', class_='btn-wrapper')
            print('\n',url)
            for tag in tags:
                anchor_tag = tag.find('a')
                # Check if the anchor tag exists
                if anchor_tag:
                    # Get the href attribute of the anchor tag
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)
                            print(f"Amazon Link: {link}")
                    # Print the Amazon link
                    print(f"Amazon Link: {link}")
                else:
                    print("Amazon link not found.")
            dict_links[url] = links
            print(links)
        elif dict_url[url] in ['hollywoodlife']:
            tags = soup.find_all('div', class_='review-product__button')
            print('\n',url)
            # print(tags)
            for tag in tags:
                anchor_tag = tag.find('a')
                # Check if the anchor tag exists
                if anchor_tag:
                    # Get the href attribute of the anchor tag
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)
                            print(f"Amazon Link: {link}")
                else:
                    print("Amazon link not found.")

            dict_links[url] = links
            print(links)
        elif dict_url[url] in ['cuisineathome', 'gardengatemagazine', 'woodsmith', 'oldhouseonline']:
            tags = soup.find_all('div', class_='product-btn-group')
            print('\n',url)
            # print(tags)
            for tag in tags:
                anchor_tag = tag.find('a')
                # Check if the anchor tag exists
                if anchor_tag:
                    # Get the href attribute of the anchor tag
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)
                            print(f"Amazon Link: {link}")
                else:
                    print("Amazon link not found.")

            dict_links[url] = links
            print(links)
        elif dict_url[url] in ['techjunkie']:
            tags = soup.find_all('div', class_='btn-three')
            print('\n',url)
            # print(tags)
            for tag in tags:
                anchor_tag = tag.find('a')
                # Check if the anchor tag exists
                if anchor_tag:
                    # Get the href attribute of the anchor tag
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)
                            print(f"Amazon Link: {link}")
                else:
                    print("Amazon link not found.")
            dict_links[url] = links
            print(links)
        elif dict_url[url] in ['usmagazine']:
            print('\n',url)
            tags = soup.select('div.article-content h3')
            link_names = [h3.find('a') for h3 in tags]
            amazon_links = [tag['href'] for tag in link_names if tag and 'amazon.com' in tag['href']]
            # Print the Amazon links
            for link in amazon_links:
                if 'amazon' in link:
                    if link not in links:
                        links.append(link)
                        print(f"Amazon Link: {link}")
                else:
                    print("Amazon link not found.")
            dict_links[url] = links
            print(links)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {str(e)}")
```

```python
# Combined code to extract headings and amazon_links

from bs4 import BeautifulSoup
import requests

dict_heading = {}
dict_links = {}

for url in Updated_URLS:
    try:
        names = []
        links = []

        page = requests.get(url, timeout=10)  # Timeout of 10 seconds
        page.raise_for_status()  # Checking for HTTP errors
        soup = BeautifulSoup(page.content, "html.parser")

        if dict_url[url] in ['thestreet', 'twentytwowords', 'mightymoms', 'sailmagazine']:
            tags = soup.find_all('div', class_='item-recommend')
            for tag in tags:
                names.append(tag.h3.text)

            links_tags = soup.find_all('div', class_='btn-wrapper')
            for tag in links_tags:
                anchor_tag = tag.find('a')
                if anchor_tag:
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)

        elif dict_url[url] in ['hollywoodlife']:
            tags = soup.find_all('article', class_='review-product')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))

            links_tags = soup.find_all('div', class_='review-product__button')
            for tag in links_tags:
                anchor_tag = tag.find('a')
                if anchor_tag:
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)

        elif dict_url[url] in ['oldhouseonline']:
            tags = soup.find_all('div', class_='product-title-wrap')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))

            links_tags = soup.find_all('div', class_='product-btn-group')
            for tag in links_tags:
                anchor_tag = tag.find('a')
                if anchor_tag:
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)

        elif dict_url[url] in ['techjunkie']:
            tags = soup.find_all('div', class_='best-product-box')
            for tag in tags:
                if tag.h2.text.replace('\n', '').replace('\t', '').replace('\xa0', '').startswith("0"):
                    names.append(tag.h2.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))

            links_tags = soup.find_all('div', class_='btn-three')
            for tag in links_tags:
                anchor_tag = tag.find('a')
                if anchor_tag:
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)

        elif dict_url[url] in ['usmagazine']:
            tags = soup.select('div.article-content h3')
            product_names = [h3.text.strip() for h3 in tags]
            for product_name in product_names:
                if '  ' in product_name:
                    names.append(product_name)

            link_names = [h3.find('a') for h3 in tags]
            amazon_links = [tag['href'] for tag in link_names if tag and 'amazon.com' in tag['href']]
            links.extend(amazon_links)

        elif dict_url[url] in ['cuisineathome', 'woodsmith', 'gardengatemagazine']:
            tags = soup.find_all('div', class_='product-title-wrap')
            for tag in tags:
                names.append(tag.h3.text.replace('\n', '').replace('\t', '').replace('\xa0', ''))

            links_tags = soup.find_all('div', class_='product-btn-group')
            for tag in links_tags:
                anchor_tag = tag.find('a')
                if anchor_tag:
                    link = anchor_tag.get('href')
                    if 'amazon' in link:
                        if link not in links:
                            links.append(link)

        print('\n', url)
        print("Product Names:")
        print(names)
        print("Amazon Links:")
        print(links)

        dict_heading[url] = names
        dict_links[url] = links

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {str(e)}")
```

```python
combined_dict = {}

# Loop through one of the dictionaries (dict_heading or dict_links) to create the combined dictionary.
for url, headings in dict_heading.items():
    if url in dict_links:
        # Ensure both dictionaries have the URL as a key
        amazon_links = dict_links[url]

        # Zip the headings and Amazon links together
        combined_data = list(zip(headings, amazon_links))

        # Assign the combined data to the URL key in the combined dictionary
        combined_dict[url] = combined_data

# Print the combined dictionary
for url, data in combined_dict.items():
    print(f"URL: {url}")
    for heading, amazon_link in data:
        print(f"   Heading: {heading}")
        print(f"   Amazon Link: {amazon_link}")
```

```python
import re

# Function to extract ASIN from an Amazon link using regex
def extract_asin(amazon_link):
    # Define a regex pattern to match ASIN in Amazon URLs
    asin_pattern = r'/dp/([A-Z0-9]{10})'

    # Use regex to search for ASIN in the Amazon link
    match = re.search(asin_pattern, amazon_link)

    # If a match is found, return the ASIN, else return None
    if match:
        return match.group(1)
    else:
        return None

# Iterate through the combined_dict and extract ASIN for each Amazon link
heading_list = []
amazon_link_list = []
asin_list = []
for url, data in combined_dict.items():
    print(f"URL: {url}")
    for heading, amazon_link in data:
        asin = extract_asin(amazon_link)
        if asin:
            print(f"   Heading: {heading}")
            heading_list.append(heading)
            print(f"   Amazon Link: {amazon_link}")
            amazon_link_list.append(amazon_link)
            print(f"   ASIN: {asin}")
            asin_list.append(asin)
        else:
            print(f"   Heading: {heading}")
            print(f"   Amazon Link: {amazon_link}")
            print("   ASIN not found")
```

```python
final_df = pd.DataFrame({'Heading':heading_list, 'Amazon_Link':amazon_link_list, 'Asin':asin_list})
final_df.head()
```
